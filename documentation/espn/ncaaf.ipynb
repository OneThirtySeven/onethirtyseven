### espn ncaa football

teams(): Returns a list of D1 college football teams

get_game(): Returns json string with game data

get_games_by_season_res(): Returns a DataFrame of game_ids for the season

___

#### Import the espn ncaaf module

```python
from onethirtyseven.espn import ncaaf
```

#### teams()

Lists of 134 D1 college football teams along with meta data including things like location, team colors and division.

```python
ncaaf_teams = ncaaf.teams()
ncaaf_teams.head()
```

___
#### get_game(game_id, stringify = False)

Parameters:

- game_id: int
- stringify (default False): bool

This returns a dict for all the game data. If stringify = True, returns json string.

*NOTE*:
Every ncaa football game has a unique **game_id**. It's listed at the top of the page.

![image.png](/img/game_id.png)

```python
game_data = ncaaf.get_game(401520169)
game_data
```
